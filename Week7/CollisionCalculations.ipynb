```python
import pandas as pd
import numpy as np

data = pd.read_csv("Data.csv")
data
```

```python
def conv_g_to_kg(df: pd.DataFrame):
    for key in df.keys():
        if(key.endswith("(g)")):
            df[f"{key[:-3]}(kg)"] = df[key] / 1000

calc_df = pd.DataFrame()
calc_df["Case"] = data["Case"]

def get_momentum(df, calc_df):
    calc_df["p1i (kgm/s)"] = df["V1i (m/s)"] * df["m1 (kg)"]
    calc_df["p2i (kgm/s)"] = df["V2i (m/s)"] * df["m2 (kg)"]
    calc_df["p1f (kgm/s)"] = df["V1f (m/s)"] * df["m1 (kg)"]
    calc_df["p2f (kgm/s)"] = df["V2f (m/s)"] * df["m2 (kg)"]
    
    calc_df["Pi (kgm/s)"] = calc_df["p1i (kgm/s)"] + calc_df["p2i (kgm/s)"]
    calc_df["Pf (kgm/s)"] = calc_df["p1f (kgm/s)"] + calc_df["p2f (kgm/s)"]

def get_KE(df, calc_df):
    def KE(m, vel):
        return 0.5 * m * (vel ** 2)
    
    calc_df["KEi (N)"] = KE(df["m1 (kg)"], df["V1i (m/s)"]) + KE(df["m2 (kg)"], df["V2i (m/s)"])
    calc_df["KEf (N)"] = KE(df["m1 (kg)"], df["V1f (m/s)"]) + KE(df["m2 (kg)"], df["V2f (m/s)"])

def get_momentum_err(df, calc_df):
    calc_df["p1f Error"] = df["m1 (kg)"] * df["verr (m/s)"]
    calc_df["p2f Error"] = df["m2 (kg)"] * df["verr (m/s)"]
    calc_df["Pf Error"] = np.sqrt(calc_df["p1f Error"] ** 2 + calc_df["p2f Error"] ** 2)

def get_KE_err(df, calc_df):
    def KE_err(m, v, v_err):
        return np.abs(m * v) * v_err
    
    KE1_err = KE_err(df["m1 (kg)"], df["V1f (m/s)"], df["verr (m/s)"])
    KE2_err = KE_err(df["m2 (kg)"], df["V2f (m/s)"], df["verr (m/s)"])
    calc_df["KEf Error"] = np.sqrt(KE1_err ** 2 + KE2_err ** 2)

def print_results(calc_df):
    for row in df.iterrows:
        print(f"Momentum ")
    
conv_g_to_kg(data)
get_momentum(data, calc_df)
get_momentum_err(data, calc_df)
get_KE(data, calc_df)
get_KE_err(data, calc_df)
display(data)
display(calc_df)
display(data.iterrows)

```
